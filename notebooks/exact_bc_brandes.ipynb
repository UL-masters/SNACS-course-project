#### Synthetic Data

**Synthetic Random Undirected Graph (n = 4902, m = 10000)**

```python
# undirected synthetic graph
rand_synth_undir = nx.read_edgelist(
                    "../data/synthetic/rand_graph_edgelist.txt",
                    nodetype=int,
                    create_using=nx.Graph()
                    )
```
